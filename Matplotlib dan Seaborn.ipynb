# DATA VISUALIZATION USING PYTHON

Python offers several plotting libraries, namely Matplotlib, Seaborn and many other such data visualization packages with different features for creating informative, customized, appealing plots to present data in the most simple and effective way.

# Matplotlib and Seaborn

Matplotlib and Seaborn are python libraries that are used for data visualization. They have inbuilt modules for plotting different graphs. While Matplotlib is used to embed graphs into applications. Seaborn is primaly used statistical graphs.

But we should we use either of two? Let's understand this with the help of a compatrative analysis. Below are the compatision between Python's two well know visualization packages Matplotlib and Seaborn.

# Matplotlib

- It used for bacis graph plotting like line charts, bar charts, etc.
- It mainly works with datasets and arrays
- Matplotlib acts productively with data arrays and frames. It regrads the aces and figure as objects
- Matplotlib is more customizable and pairs well with Pandas and Numpy for Explanatory Data Analysis

# Seaborn

- It is mainly used for statistics visualization and can perform complex visualization with fewer commands
- It works with entire datasets
- Seaborn is considerably move organized and functional than Matplotlib and treats the entire dataset as a solitary unit
- Seaborn has movie more inbuilt themes and is mainly used for statistical analysis

Import Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Line Chart

A line chart is a graph that represents information as a series of data points connected by a straight line. In line charts, each data point or marker is plotted and connected with a line or curve.

Let's consider the salary in Month (Jan-August). Let'plot a line graph using this data and see how the salary changes over time. We are going to both plot by using Matplotlib and Seaborn follows :

```python
salary=[0.8,0.9,0.5,0.5,0.8,0.5,0.3,0.6]
month=['Jan','Feb','March','April','Mei','Juni','July','August']
```

```python
plt.plot(month,salary,color='black')
plt.xlabel('Month')
plt.ylabel('Salary(Million$)')
plt.axvline(month[6],color='red',linestyle="--")
plt.axvline(month[1],color='blue',linestyle="--")
plt.title("Salary in Month (Jan-August)", fontsize=15, fontweight ='bold', color="#264653")
```

```python
salary2=[0.5,0.6,0.9,0.5,0.3,0.9,0.96,0.7]
```

```python
plt.plot(month,salary,color='black',marker='o')
plt.plot(month,salary2,color='orange',marker='x')
plt.xlabel('Month')
plt.ylabel('Salary(Million$)')
plt.title("Salary in Month (Jan-August)", fontsize=15, fontweight ='bold', color="#264653")
plt.legend(['Salary 2021','Salary 2022'], loc='lower left', fontsize=9)
```

```python
plt.figure(figsize=(16,7))
plt.plot(month,salary,color='black',marker='o')
plt.plot(month,salary2,color='orange',marker='x')
plt.xlabel('Month')
plt.ylabel('Salary(Million$)')
plt.title("Salary in Month (Jan-August)", fontsize=15, fontweight ='bold', color="#264653")
plt.legend(['Salary 2021','Salary 2022'], loc='lower left', fontsize=9)
```

```python
data_salary=pd.DataFrame({'month':month,'salary_21':salary,'salary_22':salary2})
```

```python
data_salary
```

```python
plt.figure(figsize=(10,7))
sns.set_style('whitegrid')
salary21=sns.lineplot(y='salary_21',x='month',data=data_salary,color='orange',marker='o')
salary22=sns.lineplot(y='salary_22',x='month',data=data_salary,color='red',marker='o')
plt.xlabel('Month')
plt.ylabel('Salary(Million$)')
plt.title("Salary in Month (Jan-August)", fontsize=15, fontweight ='bold', color="#264653")
plt.legend(['Salary 2021','Salary 2022'], loc='lower left', fontsize=12)
plt.show()
```

# BAR CHART

A bar chart is a type of plot represented by bars or bars, where the length of the bar is a representation of the size of a feature or variable. Bar plots usually represent the amount of data from a certain category.

```python
plt.figure(figsize=(10,7))
sns.set_style('darkgrid')
salary21=plt.bar(data_salary['month'],data_salary['salary_21'],color='green')
plt.xlabel('Month')
plt.ylabel('Salary(Million$)')
plt.title("Salary in Month (Jan-August)", fontsize=15, fontweight ='bold', color="#264653")
plt.legend(['Salary 2021','Salary 2022'], loc='upper right', fontsize=12)
```

```python
plt.figure(figsize=(10,7))
sns.set_style('darkgrid')
salary21=plt.bar(data_salary['month'],data_salary['salary_21'],color='green')
salary22=plt.bar(data_salary['month'],data_salary['salary_22'],bottom=data_salary['salary_21'],color='orange')
plt.xlabel('Month')
plt.ylabel('Salary(Million$)')
plt.title("Salary in Month (Jan-August)", fontsize=15, fontweight ='bold', color="#264653")
plt.legend(['Salary 2021','Salary 2022'], loc='upper right', fontsize=12)
```
